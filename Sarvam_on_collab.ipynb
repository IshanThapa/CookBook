```python
import ollama
import sacrebleu

# 1. Your Test Data (English -> Hindi)
test_data = [
    {"en": "Hello, how are you?", "ref": "नमस्ते, आप कैसे हैं?"},
    {"en": "I am learning AI.", "ref": "मैं एआई सीख रहा हूँ।"},
    {"en": "The weather is good today.", "ref": "आज मौसम अच्छा है।"}
]

predictions = []
references = []

print("Starting Translation Evaluation...")

for item in test_data:
    # 2. Get prediction from Sarvam-1
    # We use a clear prompt to ensure the model stays in 'translation mode'
    prompt = f"Translate the following English sentence to Hindi: {item['en']}\nHindi:"

    response = ollama.chat(model='mashriram/sarvam-1', messages=[
        {'role': 'user', 'content': prompt},
    ])

    translated_text = response['message']['content'].strip()

    predictions.append(translated_text)
    references.append([item['ref']]) # SacreBLEU expects a list of lists for references

    print(f"\nEN: {item['en']}")
    print(f"AI: {translated_text}")
    print(f"REF: {item['ref']}")

# 3. Calculate BLEU Score
# 'ja-zh' or 'intl' is often used for Indic scripts to handle characters correctly
bleu = sacrebleu.corpus_bleu(predictions, references, tokenize='intl')

print("-" * 30)
print(f"FINAL BLEU SCORE: {bleu.score:.2f}")
print("-" * 30)
```
