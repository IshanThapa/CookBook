```python
import os
from crewai import Agent, Task, Crew, Process, LLM

# 1. Setup Groq API Key (Get it from https://console.groq.com/)
os.environ["GROQ_API_KEY"] = ""

# 2. Define the Groq LLM Configuration
# llama-3.3-70b-versatile is excellent for complex research tasks
groq_llm = LLM(
    model="groq/llama-3.3-70b-versatile",
    temperature=0.2 # Lower temperature for more factual research output
)

## 3. Define Agents using the Groq LLM
researcher = Agent(
    role='Medical Research Lead',
    goal='Identify recent clinical breakthroughs in {topic}.',
    backstory='You are a healthcare AI researcher specializing in clinical validation.',
    llm=groq_llm,
    verbose=True
)

critic = Agent(
    role='Methodology Expert',
    goal='Evaluate the statistical significance of research findings.',
    backstory='You are a biostatistician with 15 years of experience in clinical trial design.',
    llm=groq_llm,
    verbose=True
)

## 4. Define Tasks (Same as before)
research_task = Task(
    description='Analyze current literature for {topic}. Focus on methodology.',
    expected_output='A bulleted summary of 3 papers with a focus on their datasets.',
    agent=researcher
)

analysis_task = Task(
    description='Review the researcher\'s findings and flag any methodology risks.',
    expected_output='A technical risk assessment report.',
    agent=critic
)

## 5. Kickoff the Crew
medical_crew = Crew(
    agents=[researcher, critic],
    tasks=[research_task, analysis_task],
    process=Process.sequential
)

result = medical_crew.kickoff(inputs={'topic': 'AI-driven sepsis early detection'})
print(result)
```
